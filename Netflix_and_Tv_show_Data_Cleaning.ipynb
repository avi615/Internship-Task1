**Checking the Null Values And Handling the Missing Values**

```python
df.isnull().sum()
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

***DATA PARSING***


converted date columns (like date_added) to datetime format using pd.to_datetime(), ensuring proper temporal analysis.

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df['date_added']
```

```python
df['date_added'].fillna(pd.Timestamp('1900-01-01'), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(20)
```

***Standardizing Text***



Stripping whitespace

---



```python
df['rating'] = df['rating'].str.strip()
df['country'] = df['country'].str.strip()
df['listed_in'] = df['listed_in'].str.strip()
```

```python
df.head(20)
```

***Feature Engineering***



Splitting and Converting Columns

```python
df[['duration_int', 'duration_type']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
df['duration_int'] = pd.to_numeric(df['duration_int'])
```
